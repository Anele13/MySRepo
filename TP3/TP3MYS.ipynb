```python
lista4 = np.random.exponential(1/4, 1000)
grafico = sns.distplot(lista4)
```

### Punto 4

De acuerdo a los resultados obtenidos en el ejercicio anterior. Especifique:

- ¿Cuál fue la media muestral de los datos generados?
- ¿Cuál fue la varianza muestral de los datos generados?
- Realice un intervalo de confianza, con  3 desvíos estándar.

#### para 100 valores:

```python
media_100_valores = round(np.mean(lista5),6)
variance_100_valores = round(np.var(lista5),6)
```

```python
print("La media muestral es: "+ str(media_100_valores))
print("La varianza muestral es: "+ str(variance_100_valores))
```

#### para 100 valores:

```python
media_1000_valores = round(np.mean(lista6),6)
variance_1000_valores = round(np.var(lista6),6)
```

```python
print("La media muestral es: "+ str(media_100_valores))
print("La varianza muestral es: "+ str(variance_100_valores))
```

```python
def intervalo_confianza(tamaño_muestra, media, desvio):
    valor_z = 2.57 #para 99% de confianza
    resul = valor_z * ((desvio) / (tamaño_muestra ** 0.5))
    return ((media-resul)*-3), ((media+resul)*3)
```

```python
print("El intervalo de confianza con +- 3 desvios para 100 valores es: " + str(intervalo_confianza(100,media_100_valores,variance_100_valores)))
print("El intervalo de confianza con +- 3 desvios para 1000 valores es: " + str(intervalo_confianza(1000,media_1000_valores,variance_1000_valores)))
```
